# make dataset for struct2graph

```python
dataset = pd.read_csv("interaction_data.csv")
```

```python
dataset = dataset.drop(columns=['interaction','mod'])
```

```python
import random
interaction = []
for idx in range(len(dataset)):
    interaction.append((dataset.iloc[idx,0],dataset.iloc[idx,1],1))
    
    found = False
    while not found:
        random_idx = random.randint(0, len(dataset) - 1)
        if dataset.iloc[random_idx]['interaction_pair'] != dataset.iloc[idx]['interaction_pair']:
            i1 = random.choice((0,1))
            i2 = random.choice((0,1))
            interaction.append((dataset.iloc[i1,0], dataset.iloc[i2,1], 0))
            found = True
```

```python
with open("Struct2Graph/interactions_data.txt", "w") as file:
    for item in interaction:
        file.write("%s\t%s\t%s\n" % item)
```

```python
all_proteins = set()
for item in interaction:
    all_proteins.add(item[0])
    all_proteins.add(item[1])
with open("Struct2Graph/list_of_prots.txt","w") as file:
    for item in all_proteins:
        file.write("%s\n" % item)
```
